```python
import math

numero = float(input("Digite um número decimal: "))

parte_inteira = int(numero)

decimal_part_abs = abs(numero - parte_inteira)

# Extract the first decimal digit
# Multiply by 10, then take the integer part to get the first digit.
first_decimal_digit = int(decimal_part_abs * 10)

rounded_number = parte_inteira

if first_decimal_digit > 5:
    if numero >= 0:
        rounded_number = parte_inteira + 1
    else: # numero < 0
        rounded_number = parte_inteira - 1
# If first_decimal_digit <= 5, rounded_number remains parte_inteira (rounds down)

print(f"O número arredondado é: {int(rounded_number)}")
```
